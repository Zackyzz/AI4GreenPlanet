```python
import pandas as pd
energy_data = pd.read_csv("Extra//energy.csv")
env_data = pd.read_csv("Extra//environment.csv")
```

```python
#energy_data = energy_data.drop(['timestamp'])

energy_data['active_power_mean'] = energy_data[["Active Power A average [kW]","Active Power B average [kW]","Active Power C average [kW]"]].mean(axis=1)
energy_data['thdi_mean'] = energy_data[["THDI A average [%]","THDI B average [%]","THDI C average [%]"]].mean(axis=1)
energy_data['thdu_mean'] = energy_data[["THDU A average [%]","THDU B average [%]","THDU C average [%]"]].mean(axis=1)
energy_data['current_average'] = energy_data[["Current A average [A]","Current B average [A]","Current C average [A]"]].mean(axis=1)
energy_data['power_mean'] = energy_data[["Power Factor A average","Power Factor B average","Power Factor C average"]].mean(axis=1)
real_data = energy_data[["power_mean","thdi_mean","current_average", "active_power_mean","thdu_mean"]]
real_data
```

```python
from sklearn.preprocessing import StandardScaler
scaler = StandardScaler()

scaled_df = pd.DataFrame(scaler.fit_transform(real_data.values), columns=real_data.columns, index=real_data.index)
scaled_df
```
